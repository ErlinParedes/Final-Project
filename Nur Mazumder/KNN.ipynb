# Understanding KNN (K nearest neighbor)

Description of KNN (K-Nearest Neighbors):

KNN is like a friendly neighbor in the world of machine learning. It helps us classify or predict things based on how close they are to other known things. This method is handy for tasks where we want to group or categorize data points.

- **Type of Problems**: KNN can tackle both classification (putting things into categories) and regression (making predictions) problems. It's often used more for classification since it assumes that similar things are close to each other.

- **Decision-Making Process**: For classification, it's like a voting system. A data point gets its label by the majority vote of its closest neighbors. If there are multiple categories, we don't need a strict 50% vote for a decisionâ€”just more votes than the others.

- **Regression Approach**: In regression (predicting values), it's similar, but instead of voting, we average the values of the nearest neighbors. This is used when dealing with continuous values.

- **Measuring Distance**: Before making decisions, we need to define how far things are from each other. We commonly use Euclidean distance (imagine the straight-line distance) but there are other methods like Manhattan, Minkowski, and Hamming distances.

KNN is versatile, making it a go-to choice when we need a flexible and intuitive approach to learning from data.


### Computing KNN: Distance Metrics

<b>Determine your distance metrics</b>

To find out which data points are near a specific query point, we calculate the distance between the query point and other data points. These distance measurements are crucial for creating decision boundaries, essentially dividing query points into distinct regions. Decision boundaries are often illustrated using Voronoi diagrams.

The most commonly used distance Metrics are listed below:

<p align="center"><b>Euclidean distance (p=2)</b>: Euclidean Distance is a popular way to measure the straight-line distance between two points. This method is commonly used for vectors with real-number values. The calculation involves finding the straight-line distance between a reference point (like a query point) and another point. The formula for Euclidean Distance helps quantify this direct distance, offering a numerical representation of how far apart these points are in a multi-dimensional space. <br /></p>

### Computing KNN: Defining the K value
<p><b>General Overview:</b><br />
The k value in the k-NN algorithm decides how many nearby neighbors to consider when determining the classification of a given point. If, for instance, k=1, the point gets classified based on its single nearest neighbor. Selecting the right k is crucial, as different values can lead to overfitting (being too specific) or underfitting (being too general). A lower k may provide high accuracy but is sensitive to noise, while a higher k might be more stable but potentially miss important patterns. It's generally advisable to use an odd k to avoid ties in classification. Cross-validation methods help in finding the most suitable k for your specific dataset.</p>

The dataset we are working with has 10 labels, meaning this dataset requires models/algorithms capable of multiclass classification.

```python
# Define a function to calculate the Euclidean distance between two points
def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b)**2))

# Implement a KNN (K-Nearest Neighbors) classifier
class KNN:
    def __init__(self, K=3):
        # Constructor to initialize the number of neighbors (K)
        self.K = K
    
    def fit(self, x_train, y_train):
        # Fit the model with training data
        self.X_train = x_train  # Training features
        self.Y_train = y_train  # Corresponding labels
        
    def predict(self, X_test):
        # Make predictions for the test data
        predictions = []  # List to store predicted labels
        for i in range(len(X_test)):
            # Calculate distances from the test instance to all training instances
            dist = np.array([euclidean_distance(X_test[i], x_t) for x_t in self.X_train])
            # Sort indices of distances and get the indices of the K-nearest neighbors
            dist_sorted = dist.argsort()[:self.K]
            # Count occurrences of each label among the K-nearest neighbors
            neigh_count = {}
            for idx in dist_sorted:
                if self.Y_train[idx] in neigh_count:
                    neigh_count[self.Y_train[idx]] += 1
                else:
                    neigh_count[self.Y_train[idx]] = 1
            # Sort the label counts in descending order and select the most frequent label
            sorted_neigh_count = sorted(neigh_count.items(), key=operator.itemgetter(1), reverse=True)
            predictions.append(sorted_neigh_count[0][0])  # Append the predicted label
        return predictions

# Evaluate the KNN model for different values of K
kVals = np.arange(3, 100, 2)
accuracies = []  # List to store accuracy scores for different K values
for k in kVals:
    model = KNN(K=k)  # Create a KNN model with the current K value
    model.fit(X_train, y_train)  # Fit the model on the training data
    pred = model.predict(X_test)  # Make predictions on the test data
    acc = accuracy_score(y_test, pred)  # Calculate accuracy
    accuracies.append(acc)  # Store accuracy for analysis
    print("K =", k, "; Accuracy:", acc)
```

From a visual view, we can see that the dataset performs as it's best when k = 3, 5 and 7 and 9, beyond that, the performance diminishes gradually, meaning the lowest k value yields the highest performance.

```python
max_index = accuracies.index(max(accuracies))   # we can also take the maximum value of the accuracies array and find the corresponding index
print(max_index)  # output should be 0
```

```python
# visualize the graph
from matplotlib import pyplot as plt 
plt.plot(kVals, accuracies) 
plt.xlabel("K Value") 
plt.ylabel("Accuracy")

# the line graph gives us a visual representation of the above iterations
```
